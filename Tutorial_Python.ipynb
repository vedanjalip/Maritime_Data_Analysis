Lists and dictionaries

```python
# Creating a list of numbers
numbers = [1, 2, 3, 4, 5]            # Index ranges from 0 to (number of elements - 1)

# Accessing elements of the list
print("First element:", numbers[0])  # First element
print("Last element:", numbers[-1])  # Last element
#print("Last element:", numbers[4])  # Another way: Last element


# Adding elements to the list
numbers.append(6)
print("After appending:", numbers)

# Removing elements from the list
numbers.remove(3)
print("After removing 3:", numbers)

# Slicing the list
print("First three elements:", numbers[:3])


# Creating a dictionary
person = {
    'name': 'Alice',
    'age': 25,
    'city': 'New York'
}
print(person)
# Accessing values using keys
print("Name:", person['name'])
print("Age:", person['age'])

# Adding a new key-value pair
person['email'] = 'alice@example.com'
print("After adding email:", person)

# Updating a value
person['age'] = 26
print("Updated age:", person)

# Removing a key-value pair
del person['city']
print("After deleting city:", person)
```

Arrays

```python
import numpy as np

# Creating a NumPy array
array = np.array([1, 2, 3, 4, 5])

# Accessing elements
print("First element:", array[0])

# Performing arithmetic operations
print("Array multiplied by 2:", array * 2)

# Creating arrays of zeros and ones
zeros = np.zeros(5)
ones = np.ones(5)

print("Array of zeros:", zeros)
print("Array of ones:", ones)

# Reshaping arrays
matrix = np.arange(1, 10).reshape(3, 3)
print("Reshaped array:\n", matrix)
```

Plotting data using matplotlib

```python
import matplotlib.pyplot as plt

# Creating sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

# Line plot
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# Bar plot
plt.bar(x, y)
plt.title("Bar Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# Scatter plot
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Data Analysis with Pandas

```python
import pandas as pd

# Creating a Pandas DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}

df = pd.DataFrame(data)

# Displaying the DataFrame
print(df)

# Accessing columns
print("\n","Names:", df['Name'])

# Adding a new column
df['Salary'] = [50000, 60000, 70000]
print("After adding Salary column:\n", df)

# Filtering rows
print("People older than 30:\n", df[df['Age'] > 30])
```
